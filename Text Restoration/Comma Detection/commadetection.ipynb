```python
# import h2o and all algorithms to be used
import h2o
from h2o.estimators import H2OGradientBoostingEstimator
# initialise h2o
h2o.init()
```

```python
# importing the training/testing dataset
dataset = 'emma_pos_commadetection_in.txt'
df = h2o.import_file(dataset)

# setting the feature and response columns
y = 'C17'
df[y] = df[y].asfactor()
x = df.columns
x.remove(y)
```

```python
# importing the story that is to be used to make predictions on
pred_dataset = 'prediction_commadetection_in.txt'
pred_df = h2o.import_file(pred_dataset)
```

```python
# Defining the gbm algorithm that is to be used
gbm = H2OGradientBoostingEstimator(ntrees=50,       # number of trees used
                                   max_depth=10,    # how deep each tree goes
                                   min_rows=2,     # how many rows each tree has
                                   learn_rate=0.2)  # learn rate of algorithm

# Fitting the gbm algorithm (training it)
gbm.train(x = x,                    # feature columns
          y = y,                    # response column
          training_frame = df)      # entire training frame
```

```python
# showing stats that tell us how well the algorithm did
# h2o automatically test so these stats are on the testing set
perf = gbm.model_performance()
print(perf)
```

```python
# making predictions with the gbm algorithm
# storing predictions in variable
preddrf = gbm.predict(pred_df)

# binding predictions together as list containing a list for each predictions
all = preddrf["predict"].cbind(pred_df).as_data_frame(use_pandas = False)

del all[0]              # deleting the first list (contains column names)
for item in all:        # making the predictions from floats to integers
    item[0] = int(item[0])

# open a new file where we store the output text
new = open("new_emmapos_commadetection_gbm_out.txt", "w+")

# writing the output text into the new file
for item in all:
    # if prediction is '0', write item + space
    if item[0] == 0:
        new.write(item[-1])
        new.write(" ")
    # if prediction is '1', write item + ',' + space
    if item[0] == 1:
        new.write(item[-1])
        new.write(", ")

# Closing the new file
new.close()
```

```python
# shutting down the h2o cluster
h2o.cluster().shutdown()
```

```python
# the output from above still includes the POS-tags which need to be removed
#opening the file and storing content in string variable
with open('new_emmapos_commadetection_gbm_out.txt', 'r') as file:
    data = file.read().replace('\n', ' ')

# splitting the string at the spaces and storing the output in variable 
data = data.split(' ')
text = ''

# running through each item in the list
for i in range(len(data)):
    for item in data:
        if item[-1] == '>': # checking whether item contains POS tag
            for char in item:
                if char.isalpha() and char == char.lower():
                    text += char # adding character from item to text variable
            text += ' '
        if item[-1] != '>':
            for char in item: # checking whether item contains inserted comma
                if char.isalpha() and char == char.lower():
                    text += char # adding character from item to text variable
            text += item[-1]
            text += ' '
        # accounting for all punctuation marks in the story
        if item == '.<.>':
            text += '. '
        if item == '-<.>':
            text += '-'
        if item == ';<.>':
            text += '; '
        if item == '(<.>':
            text += '('
        if item == ')<.>':
            text += ')'

# splitting the text again
data = text.split(' ')

# dealing with remaining spaces
full_text = ''
for item in data:
    if item:
        if item.isalpha():
            full_text += item
            full_text += ' '
            continue
        elif item[1:].isalpha():
            full_text += item[0]
            full_text += item[1:]
            full_text += ' '
        elif item[:-1].isalpha():
            full_text += item[:-1]
            full_text += ' '
            full_text += item[-1]
            full_text += ' '
        elif item == '.':
            full_text += item + ' '
        elif item == ',':
            full_text += item + ' '
        elif item == '-':
            full_text += item + ' '
        elif item == '(':
            full_text += item + ' '
        elif item == ')':
            full_text += item + ' '
        elif item == ';':
            full_text += item + ' '
            
# opening a new file
new = open("new_bigcommadetection_gbm_out_finalfinal.txt", "w+")

# storing full_text in the file character by character
for char in full_text:
    new.write(char)

# closing the new file
new.close()

# the error can be ignored, the output file is correct
```
